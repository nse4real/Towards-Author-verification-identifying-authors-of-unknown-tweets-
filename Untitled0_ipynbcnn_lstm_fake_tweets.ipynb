```python
## preprocessing functions
def remove_URL(tweet):
    """Remove URLs from a sample string"""
    return re.sub(r"http\S+", "", tweet)

def remove_numbers(tweet):
  """Remove words with numbers"""
  return re.sub("[0-9]","",tweet)   

def remove_non_ascii(tweet):
    """Remove non-ASCII characters from list of tokenized words"""
    return unicodedata.normalize('NFKD', tweet.encode('ascii', 'ignore').decode('utf-8', 'ignore'))
```

```python
# Use apply function to apply preprocessing
df2["tweet"] = df2["tweet"].apply(lambda t: remove_URL(t)) #remove urls
df2["tweet"] = df2["tweet"].apply(lambda t: remove_numbers(t))
df2["tweet"] = df2["tweet"].apply(lambda t: remove_non_ascii(t))
print(df2.shape)
```

#### Data preprocessing

```python
## C0nvert tweet_device and username to categorical type
df2["tweet_device"] = df2["tweet_device"].astype('category')
df2["username"] = df["username"].astype('category')
df2["tweet_device_code"] = df2["tweet_device"].cat.codes
df2["username_code"] = df2["username"].cat.codes
print(df2.shape)
df2.head()
```

```python
#Dividing the dataset into features and lables
tweets = df2['tweet']
labels = df2['username_code']

#check the distribution of lebels
classnames = np.array(df2['username'].cat.categories)
print(classnames)
labels_count = labels.value_counts()
print(labels_count)

plt.bar(classnames,labels_count, color ='blue', width = 0.4)

plt.xticks(rotation=60)
plt.xlabel("Authors")
plt.ylabel("No. of tweets")
plt.title("Number of tweets per author")
plt.show()
```

```python
tweets_split = []

tkr = RegexpTokenizer(r"\w+")

for i, line in enumerate(tweets):
    #print(line)
    tweet = tkr.tokenize(line)
    tweets_split.append(tweet)

print(tweets_split[1])
```

```python
gn_vec_zip_path = "/content/drive/MyDrive/Colab Notebooks/GoogleNews-vectors-negative300.bin.gz"
gn_vec_path = "/content/drive/MyDrive/Colab Notebooks/GoogleNews-vectors-negative300.bin"
#Extracting the required model
with gzip.open(gn_vec_zip_path, 'rb') as f_in:
    with open(gn_vec_path, 'wb') as f_out:
        shutil.copyfileobj(f_in, f_out)
   
print(f"Model at {gn_vec_path}")
```

```python
# prepare target
def prepare_targets(labels):
	label_binarizer = LabelBinarizer()
	label_binarizer.fit(range(max(labels)+1))
	return label_binarizer.transform(labels)
```

```python
'''
https://colab.research.google.com/github/practical-nlp/practical-nlp/blob/master/Ch3/05_Pre_Trained_Word_Embeddings.ipynb?authuser=1
Use pretrained Word2Vec model from google but trim the word list to 50,000 compared to 3,000,000 in the original
Google pretrained model
'''
pretrainedpath = gn_vec_path
w2vModel = word2vec.KeyedVectors.load_word2vec_format(pretrainedpath, binary=True, limit=50000)

#Convert words to integers
tokenizer = Tokenizer()
tokenizer.fit_on_texts(tweets_split)
X = tokenizer.texts_to_sequences(tweets_split)

#length of tweet to consider
maxlentweet = 30
#add padding
X = pad_sequences(X, maxlen=maxlentweet)
print(X.shape)
```

```python
X[:10,]
```

```python
def init_model():
  #create a embedding layer using Google pre triained word2vec (50000 words)
  embedding_layer = Embedding(input_dim=w2vModel.syn0.shape[0], output_dim=w2vModel.syn0.shape[1], weights=[w2vModel.syn0], 
                              input_length=X.shape[1])

  #create model combining LSTM with 1D Convonutional layer and MaxPool layer
  lstm_out = 150

  model = Sequential()
  model.add(embedding_layer)
  model.add(Conv1D(filters=64, kernel_size=5, activation='relu', padding='causal'))
  model.add(MaxPooling1D(pool_size=2))
  model.add(Dropout(0.7))
  model.add(LSTM(units=lstm_out))
  model.add(Dropout(0.7))
  model.add(Flatten())
  model.add(Dense(6, activation='softmax'))
  model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
  print(model.summary())
  return model
```

### Split data into train, test and validation

```python
#Here we split our data into train and test
training_data, testing_data, training_labels, testing_labels = train_test_split(X, labels, test_size=0.2, random_state=34)
print("The dimension of training and test dataset are:",training_data.shape,testing_data.shape)
```

```python
#split training data into train and test such that test represent the validation dataset
X_train, X_test, Y_train, Y_test = train_test_split(training_data, training_labels, test_size= 0.1, random_state = 24)

y_train_enc = prepare_targets(Y_train)
y_test_enc = prepare_targets(Y_test)
print(X_train.shape,y_train_enc.shape)
```

### Training the model

```python
# Open all files for writing everytime the program is ran from the beginning
filenames = ['cnn-lstm-output.csv','test-cnn-lstm-output.csv']
#Header
header = ['Experiment-Number','Loss','Accuracy','Total-F1','cm']
for outputFilename in filenames:
  # open the file using open() function
  with open(path+outputFilename, 'w', newline='') as f:
      writer = csv.writer(f)
      writer.writerow(header)
```

```python
def perform_experiment(experimentNumber, outputFilename):
  print(f"PERFORMING EXPERIMENT {experimentNumber}\n\n")
  model = init_model();
  #fit model
  batch_size = 32
  model.fit(X_train, y_train_enc, epochs=5, verbose=1, batch_size=batch_size)

  #analyze training results
  loss, acc = model.evaluate(X_test, y_test_enc, verbose = 2, batch_size=batch_size)
  #y_pred = model.predict(X_test)
  print("TRAINING RESULTS\n\n")
  preds = model.predict(X_test, verbose=0)
  predictions = preds.argmax(axis=1)
  print(classification_report(Y_test, predictions))
  cm = confusion_matrix(Y_test, predictions)
  print(cm)
  print(f"Loss: {loss}")
  print(f"Accuracy: {acc}")
  f1s = f1_score(Y_test, predictions, average=None)
  total_f1s = np.sum(f1s)
  print("Total f1 over classes: {}".format(total_f1s))
  turple = [experimentNumber, loss, acc, total_f1s,cm.reshape([1,36]).tolist()]
  #print("****************************************************************")
  #print("EXPERIMENT VALIDATION RESULT")
  with open(path+outputFilename, 'a+', newline='') as f:
      writer = csv.writer(f)
      writer.writerow(turple)

  #perform evalation with test data
  test_enc = prepare_targets(testing_labels)
  print("TEST RESULTS\n\n")
  preds = model.predict(testing_data, verbose=0)
  predictions = preds.argmax(axis=1)
  print(classification_report(testing_labels, predictions))
  cm = confusion_matrix(testing_labels, predictions)
  print(cm)
  acc = accuracy_score(testing_labels, predictions)
  print(f"Accuracy: {acc}")
  f1s = f1_score(testing_labels, predictions, average=None)
  total_f1s = np.sum(f1s)
  print("Total f1 over classes: {}".format(total_f1s))
  turple = [experimentNumber, 0, acc, total_f1s,cm.reshape([1,36]).tolist()]
  #print("****************************************************************")
  #print("EXPERIMENT TEST RESULT")
  name = path+"test-"+outputFilename
  with open(name, 'a+', newline='') as f:
      writer = csv.writer(f)
      writer.writerow(turple)
     

perform_experiment(1,filenames[0])
perform_experiment(2,filenames[0])
perform_experiment(3,filenames[0])
perform_experiment(4,filenames[0])
perform_experiment(5,filenames[0])
```

```python
def convert_cm_to_confusion_matrix(group_cm):
  cm_lists = []
  for cm in group_cm:
    #print(cm)
    num = []
    ch_list = []
    index = 0
    while index < len(cm):
      if cm[index].isdigit():
        num.append(cm[index])
      else:
        if len(num)>0:
          ch_list.append(int("".join(num)))
          num = []
      index+=1
    #print(ch_list)
    cm_lists.append(ch_list)
  final_cm = np.array(cm_lists)
  #print(final_cm)
  f = np.sum(final_cm, axis=0)
  #print(f)
  return f

## group the data by classifier-code and split-number
grouped_df = output_csv_df.groupby(['Experiment-Number'])
experiments_result = {
    'Experiment-Number':[],
    'Loss':[],
    'Accuracy': [],
    'Total-F1': []
}
for i in range(0,36):
  experiments_result[i] = []

for name, group in grouped_df:
  experiments_result['Experiment-Number'].append(name)
  experiments_result['Loss'].append(np.mean(group['Loss']))
  experiments_result['Accuracy'].append(np.mean(group['Accuracy']))
  experiments_result['Total-F1'].append(np.mean(group['Total-F1']))

  cm = convert_cm_to_confusion_matrix(group['cm'])
  for i in range(0,36):
    experiments_result[i].append(cm[i])
  
print("TRAINING RESULT FOR EACH EXPERIMENT")
df = pd.DataFrame(experiments_result)
display(df)
```

#### Precision, recall and confusion matrix

```python
from tabulate import tabulate
%matplotlib inline
import seaborn as sns

class_labels = ['Shteyngart', 'askanyone', 'harlancoben ', 'megcabot','paulocoelho', 'tejucole']

def print_precision_recall(cf_matrix):
  metrics = {"username":class_labels,"precision":[],"recall":[]}
  row_sum = np.sum(cf_matrix,axis=1)
  col_sum = np.sum(cf_matrix, axis=0)
  for i in range(len(class_labels)):
    metrics['precision'].append(cf_matrix[i,i]/row_sum[i]) 
    metrics['recall'].append(cf_matrix[i,i]/col_sum[i]) 
  print(tabulate(metrics,  headers=["Username","Precision","Recall"])) 

def plot_confusion_matrix(df_row):
  matrix_values = np.round_(np.array(df_row.iloc[list(range(3,39))],dtype='i'))
  cf_matrix = matrix_values.reshape(6, 6)

  print(f'Precision and recall for CNN LSTM\n\n')
  print_precision_recall(cf_matrix)

  ax = sns.heatmap(cf_matrix, fmt='g', annot=True, cmap='Blues')
  ax.set_title(f'Seaborn Confusion Matrix for CNN LSTM\n\n');
  ax.set_xlabel('\nActual Values')
  ax.set_ylabel('Predicted Values\n');
  ax.xaxis.set_ticklabels(class_labels,rotation=45)
  ax.yaxis.set_ticklabels(class_labels,rotation=45)
  ## Display the visualization of the Confusion Matrix.
  plt.show()

plot_confusion_matrix(df2)
```
